```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
base = "log2/"
```

```python
def processFile(path):
    path = base+path
    p = float(int(path.split(".")[0].split("p")[-1]) / 10)
#     print("percentage = ", p)
    rvals =[]
    f = open(path, "r")
    for line in f:
        if "average mirgate ratio = " in line:
            #Perturbations: 58.215 ms; average mirgate ratio = 0.406535; rebuild false
            ratio = float(line.split(" =")[-1].split(";")[0])
            rvals.append(ratio)
    return p, rvals
```

```python
plt.rcParams["figure.figsize"] = (20,10)
p, ratios = processFile("i10_p0.log")
plt.plot(ratios,'r--', label = "p = "+str(p))
print(p, ratios)
p, ratios = processFile("i10_p4.log")
plt.plot(ratios, label = "p = "+str(p))
print(p, ratios)
p, ratios = processFile("i10_p5.log")
plt.plot(ratios, label = "p = "+str(p))
print(p, ratios)
p, ratios = processFile("i10_p6.log")
plt.plot(ratios, label = "p = "+str(p))
print(p, ratios)
p, ratios = processFile("i10_p7.log")
plt.plot(ratios, label = "p = "+str(p))
print(p, ratios)
p, ratios = processFile("i10_p10.log")
plt.plot(ratios, '--', label = "p = "+str(p))
print(p, ratios)
plt.legend()
plt.title("Migrate ratio")
plt.grid(True, axis='both')
plt.show()
```
